```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from Project1 import LogisticRegression, ClassificationMetrics
```

```python
data = datasets.load_breast_cancer()
X, y = data['data'], np.expand_dims(data['target'], axis=-1)
X.shape, y.shape
```

```python
LR = LogisticRegression()
```

```python
LR.fit(X=X, y=y, num_iterations=3000, learning_rate=0.5, verbose=True)
```

```python
y_hat = LR.predict(X)
```

```python
CM = ClassificationMetrics(y, y_hat)
CM.accuracy(), CM.precision(), CM.recall(), CM.F1()
```
